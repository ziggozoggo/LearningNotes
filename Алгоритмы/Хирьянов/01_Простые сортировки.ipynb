```python
def insert_sort(unsorted_list):
    """ Сортировка списка unsorted_list вставками
    Изменяет переданный список
    """
    len_unsorted_list = len(unsorted_list)
    for top in range(1, len_unsorted_list):
        # Двигаемся справа налево по списку с 
        # индексом объекта top и сравниваем его с другими элементами
        k = top
        
        # До тех пор, пока индекс k больше нуля и 
        # предыдущий элемент больше текущего top
        # совершаем обмен значениями между соседними элементами списка
        
        # В условии unsorted_list[k-1] не попадем на -1 индекс, т.к. при k = 0
        # цикл прекратится без вычисления второго условия
        while k > 0 and unsorted_list[k-1] > unsorted_list[k]:
            # обмен значений местами
            unsorted_list[k], unsorted_list[k-1] = unsorted_list[k-1], unsorted_list[k]
            k -= 1

def choise_sort(unsorted_list):
    """ Сортировка списка unsorted_list методом выбора
    Изменяет переданный список
    """
    len_unsorted_list = len(unsorted_list)
    # Цикл до N-1 т.к. N-ый элемент автоматически встанет на свою позицию
    for pos in range(0, len_unsorted_list-1):
        # Анализ элементов правее отсортированного pos
        for k in range(pos+1, len_unsorted_list):
            if unsorted_list[k] < unsorted_list[pos]:
                unsorted_list[k], unsorted_list[pos] = unsorted_list[pos], unsorted_list[k]

def bubble_sort(unsorted_list):
    """Сортировка списка unsorted_list методом пузырька
    Изменяет переданный список
    """
    # Отличие от предыдущей заключается в учете количества проходов
    N = len(unsorted_list)
    for bypass in range(1, N):
        for k in range(0, N-bypass):
            if unsorted_list[k] > unsorted_list[k+1]:
                unsorted_list[k], unsorted_list[k+1] = unsorted_list[k+1], unsorted_list[k]
        
    
```

### Сортировка подсчётом O(n)

```python
def count_sort(unsorted_list):
    """Сортировка списка unsorted_list методом подсчёта
    """
    # Хранилище для частотного анализа содержимого списка
    # Для его формирования необходимо знать максимум списка
    max_value = 0
    for value in unsorted_list:
        if value > max_value:
            max_value = value

    freq_storage = [0] * (max_value + 1) # На один больше, т.к. счёт начинаем с нуля
    sorted_list = []
    
    # Подсчёт различных элементов:
    for number in unsorted_list:
        freq_storage[number] += 1
    
    unsorted_list = []
    for number, counter in enumerate(freq_storage):
        if counter != 0:
            sorted_list = sorted_list + [number] * counter
    
    return sorted_list
```

```python
# тестирование
def test_sort(sort_algorithm_function):
    print("Тестируем: ", sort_algorithm_function.__doc__)
    print("testcase #1:", end="")
    unsorted_list = [4, 2, 5, 1, 3]
    sorted_list = [1, 2, 3, 4, 5]
    
    sort_algorithm_function(unsorted_list)
    print("Ok" if unsorted_list == sorted_list else "Fail")
    
    print("testcase #2:", end="")
    unsorted_list = list(range(10,20)) + list(range(0, 10))
    sorted_list = list(range(0, 20))
    
    sort_algorithm_function(unsorted_list)
    print("Ok" if unsorted_list == sorted_list else "Fail")
    
    print("testcase #3:", end="")
    unsorted_list = [4, 2, 4, 5, 1, 1, 3]
    sorted_list = [1, 1, 2, 3, 4, 4, 5]
    
    sort_algorithm_function(unsorted_list)
    print("Ok" if unsorted_list == sorted_list else "Fail")
    print()
    
def test_sort_count(sort_algorithm_function):
    print("Тестируем: ", sort_algorithm_function.__doc__)
    print("testcase #1:", end="")
    unsorted_list = [4, 2, 5, 1, 3]
    sorted_list = [1, 2, 3, 4, 5]
    
    res = sort_algorithm_function(unsorted_list)
    print("Ok" if res == sorted_list else "Fail")
    
    print("testcase #2:", end="")
    unsorted_list = list(range(10,20)) + list(range(0, 10))
    sorted_list = list(range(0, 20))
    
    res = sort_algorithm_function(unsorted_list)
    print("Ok" if res == sorted_list else "Fail")
    
    print("testcase #3:", end="")
    unsorted_list = [4, 2, 4, 5, 1, 1, 3]
    sorted_list = [1, 1, 2, 3, 4, 4, 5]
    
    res = sort_algorithm_function(unsorted_list)
    print("Ok" if res == sorted_list else "Fail")
    print()
```

```python
test_sort(insert_sort)
test_sort(choise_sort)
test_sort(bubble_sort)
test_sort_count(count_sort)
```
